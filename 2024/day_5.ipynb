```python
import numpy as np
```

```python
with open("./data/input_day5.txt", "r") as f:
    data = f.readlines()
```

```python
# separate between the instructions and the number list
executions = []
instructions = []
for line in data:
    line = line[:-1] # remove newline marker
    if line.find(",") > 0:
        line_list = [int(num) for num in line.split(',')]
        executions.append(line_list)
    elif line.find("|") > 0:
        inst_nums = [int(num) for num in line.split("|")]
        instructions.append(inst_nums)
    else:
        pass

instructions = np.array(instructions)
```

```python
sum_of_middle_nums = 0
# for each execution line
for iexec, exec in enumerate(executions):
    # iterate over each number and check if 
    # the rules apply
    rule_corr_applied = True
    for inum, num in enumerate(exec):
        # if the number is within the `before` column
        indices_num_in_before = np.argwhere(num == instructions[:,0]).T[0]

        # check if the subsequent numbers after that number are in 
        # the `after` part of the instructions
        # if all numbers are not in there we terminate the process
        if not np.all(np.isin(exec[inum+1:], instructions[indices_num_in_before, 1])):
            rule_corr_applied = False
            break

    # if the rule is correctly applied, then find the middle number
    if rule_corr_applied:
        sum_of_middle_nums += exec[len(exec)//2]
            
```

```python
print(f"sum of all middle numbers with correctly applied rules: {sum_of_middle_nums}")
```
